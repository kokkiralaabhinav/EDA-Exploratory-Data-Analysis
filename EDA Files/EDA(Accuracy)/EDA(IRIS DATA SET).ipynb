### **Scaling or Normalisation - It done only for inputs(Scaling the number of digits in difference input columns)**

**Before Normalisation:**

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)  # My random values, ones generasted will remains constant
x1 = np.random.randint(1,50,30)
x1 = np.sort(x1) #it sorts array in acending order
x2 = np.random.randint(10,70,30)
plt.plot(x1,x2)
```

**After Normalistaion**

```python
# These graphs are no whererelated to machine learning
# Here we are only understanding the concept of scaling or Normalisation
x1min = min(x1)
x1max = max(x1)
x2min = min(x2)
x2max = max(x2)
x1norm = (x1-x1min)/(x1max-x1min)
x2norm = (x2-x2min)/(x2max-x2min)
plt.plot(x1norm,x2norm) 
```

### **Train Test and split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0)
```

# **Applying a Classifier/Regressor/Cluster :**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model
```

### **Model fitting**

```python
model.fit(x_train,y_train) #using the training data, we train our model
```

### **Predict the Output**

```python
y_pred = model.predict(x_test)
y_pred #predicted values
```
